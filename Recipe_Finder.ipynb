```python
import ast

def convert_to_list(ner_string):
    return ast.literal_eval(ner_string)

df['NER'] = df['NER'].apply(convert_to_list)
```

```python
df['NER'] = df['NER'].apply(lambda x: ' '.join(x))
```

```python
input_str = input('Enter ingredients:')
```

```python
df['NER'][2]
```

```python
print(input_str)
input_str = input_str.replace('"', '').replace(',', ' ').strip()
input_str = input_str.lower()
print(input_str)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
vectorizer = TfidfVectorizer()
all = df['NER'].tolist() + [input_str]
```

```python
tfidf_matrix = vectorizer.fit_transform(all)
cosine_sim = cosine_similarity(tfidf_matrix[-1], tfidf_matrix[:-1])
```

```python

top_indices = cosine_sim.argsort()[0][-3:][::-1]

print("Top 5 similar recipes:")
for index in top_indices:
    print(f"Recipe: {df.iloc[index]['title']}, Similarity: {cosine_sim[0][index]}")
    print(f"Ingredients: {df.iloc[index]['ingredients']}")
    print(f"Instructions: {df.iloc[index]['directions']}")
    print()
```
